```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('churn.csv')
df
```

```python
sns.set_style(style='whitegrid')
plt.figure(figsize=(12,10))
```

```python
sns.countplot(x='Exited', data=df)
plt.title('dist of churn')
```

```python
sns.histplot(data=df, x='Age', kde=True)
plt.title('age dist')
```

```python
sns.scatterplot(data=df, x='CreditScore', y= 'Age', hue='Exited')
plt.title('credit score vs age')
```

```python
sns.boxplot(data=df, x='Exited', y= 'Balance')
plt.title('balance dist by exited')
```

```python
sns.boxplot(data=df, x= 'Exited', y= 'CreditScore')
plt.title('credit dist by exited')
```
